```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./schedule.csv')
df
```

```python
lose_total = {'中信兄弟': 0, '樂天桃猿': 0, '統一獅': 0, '味全龍': 0, '富邦悍將': 0}
win_total = {'中信兄弟': 0, '樂天桃猿': 0, '統一獅': 0, '味全龍': 0, '富邦悍將': 0}
standings = []
for i in range(0, 300):
    game = df.iloc[i]
    # Add `total`
    if game['team_lose'] != 'Draw':
        lose_total[game['team_lose']] += 1
    if game['team_win'] != 'Draw':
        win_total[game['team_win']] += 1

    # Calculate the winning rate
    temp_standings = {'中信兄弟': 0.0, '樂天桃猿': 0.0, '統一獅': 0.0, '味全龍': 0.0, '富邦悍將': 0.0}
    for key in temp_standings.keys():
        try:
            temp_standings[key] = round(win_total[key]/(lose_total[key] + win_total[key]), 2)
        except:
            temp_standings[key] = -1.0

    temp_standings = sorted(temp_standings.items(), key=lambda x:-x[1])
    standings.append(temp_standings)
    # print(temp_standings)
```

```python
A_ranked_data = {'中信兄弟': [], '樂天桃猿': [], '統一獅': [], '味全龍': [], '富邦悍將': []}
for game in standings:
    rank = 0
    count = 0
    prev_score = None
    for team, score in game:
        if score != -1.0:
            if score != prev_score:
                rank += 1 + count
                count = 0
            else:
                count += 1
            A_ranked_data[team].append(rank)
        else:
            A_ranked_data[team].append(0)
        prev_score = score
```
